```python
import seaborn as sns #package for plotting our data
import matplotlib.pyplot as plt #package for plotting our data
import numpy as np #package for integrations
from google.colab import files #For uploading and downloading files
import pandas as pd #for data analysis
import io #for reading uploaded data
from dataclasses import dataclass #to use special methods for our class
```

```python
class LCMeasurement:
  #creating class for measurement of P_s (spontaneous polarization) for our Liquid Crystal

  def __init__(self, fname: str, df: pd.DataFrame, temp: float, lc_name: str):
    self.fname = fname
    self.df = df
    self.temp = temp
    self.lc_name = lc_name

  @classmethod
  def from_csv(cls, fname, content=None):
      if content:
        df = pd.read_csv(io.BytesIO(content)) #Upload file from computer
      else:
        df = pd.read_csv(fname) #Using the file uploaded earlier
      temp = float(fname.split('_', 1)[0])
      lc_name = fname.split('_', 7)[5]
      return cls(fname, df, temp, lc_name)


def plot_raw_signal(self):
    sns.scatterplot(data=self.df, x='Time (s)', y='Channel 3 (VOLT)', alpha=0.7)
    plt.xlabel('Time (s)')
    plt.ylabel('Response (V)')
    plt.legend(title=f'T(C): {self.temp}')
    plt.title(f"Raw Signal of {self.lc_name}")
    plt.show()
```

```python
uploaded = files.upload()
fname, content = next(iter(uploaded.items()))
```

```python
measurement = LCMeasurement.from_csv(fname, content)

# Plot raw signal
measurement.plot_raw_signal()
```
